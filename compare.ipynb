```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
import pandas as pd
```

```python
# creat Dkl formula for p & q Dkl=sum(p * log(p / q)) 
def kl(p, q):
    result = np.sum(np.where(np.logical_and(p !=0, q!=0), p * np.log(p / q), 0))
    return result
```

```python
data = np.arange(-10, 10, 0.001)
```

```python
#design p & q with pdf(x, loc=?, scale=?)
p = norm.pdf(data, 0, 2)
q = norm.pdf(data, 2, 3)
#make p & q plot
plt.plot(data, p, c='blue')
plt.plot(data, q, c='red')
```

```python
kl(p, q)
```

```python
#design p & q with pdf(x, loc=?, scale=?)
p = norm.pdf(data, 0, 2)
q = norm.pdf(data, 0.2, 2.6)
#make p & q plot
plt.plot(data, p, c='blue')
plt.plot(data, q, c='red')
```

```python
kl(p, q)
```
