```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.random.randint(low= 1, high= 100, size= (10, 10))
```

```python
print(data)
```

```python
#Basic Heatmap
hm_01 = sns.heatmap(data = data)
```

```python
#Anchoring the colormap
hm_02 = sns.heatmap(data= data,
                   vmin= 30,
                   vmax= 70)
```

```python
#Choosing the colormap
hm_03 = sns.heatmap(data= data, cmap='tab20')
```

```python
#Choosing the colormap
hm_03 = sns.heatmap(data= data, cmap='Accent')
```

```python
#Centering the colormap
hm_04 = sns.heatmap(data= data, cmap='tab20', center= 0)
```

```python
#Displaying the cell values with annot=True
hm_05 = sns.heatmap(data= data, cmap= 'tab20', annot=True)
```

```python
hm_05 = sns.heatmap(data= data, cmap='tab20', vmin= 30, vmax=85, annot=True)
```

```python
#Customizing the separating line
hm_06 = sns.heatmap(data= data, cmap= 'Accent', annot= True, linewidth= 3, linecolor= 'yellow')
```

```python
#Hiding the colorbar
#We can disable the colorbar by setting the cbar parameter to False.
hm_07 = sns.heatmap(data= data, cmap='tab20', cbar= False, annot= True)
```

```python
#Removing the labels
#We can disable the x-label and the y-label by passing False in the xticklabels and yticklabels parameters respectively
hm_05 = sns.heatmap(data= data, cmap='tab20', annot=True, xticklabels=False, yticklabels= False, cbar= False)
```
